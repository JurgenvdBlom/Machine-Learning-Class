```python
import re #import regular expressions

text="ababbfmvkbmrwvlek"
match=re.search("ab",text)
print(match.group())

# ^\w[abc] find a start position of a string with any word character and directly after either a, b, or c
# if we add $ it should also end after this a,b,c
# \b means for boundaries 
# a.a matches a (any) a 
# a.?a can match 0 or 1 character in between
# a.+a can match at least one (or more) character in between two a
# a.*a matches all a a with nothing or something in between

#(ya){2} where it happens two times in a row, so yaya
 
```

```python
# (ab)\1  ab and ab 1 more time; so looks for abab
# \b\d{2,3} so start at a boundary and with a digit with length 2 to 3 (for phone numbers that start with e.g. 01 or 010)
#\b\d{2,3}[- ]?\d{3,4}[- ]?\d{4}\b To look for phone numbers with - or space or nothing in between pairs e.g. 010-3435-034

#now use grouping for this phone number look up:
#(\b\d{2,3})[- ]?(\d{3,4})[- ]?(\d{4}\b)
# and then if you use \1 you refer to the first group (\b\d{2,3})
# so e.g. \1-****-\3 replaces the second group with -****-
# if you use for the second group (?: ...) instead, tis group won't be captured, so instead you should put
# so e.g. \1-****-\2 replaces the non-captured group with -****-
```

### use [Regular Expression Generator](https://regex101.com/) to try out the reg expression and convert it to python code, like done below!

```python
# coding=utf8
# the above tag defines encoding for this document and is for Python 2.x compatibility

import re

regex = r"(\b\d{2,3})[- ]?\d{3,4}[- ]?(\d{4})\b"

test_str = ("hello world\n"
	"4aa  13\n"
	"aba  b44\n"
	"ababaa5\n"
	"yaya---\n"
	"ikwak2@cau.ac.kr\n"
	"il.youp.kwak@gmail.com\n"
	"02-111-1234\n"
	"010-3234-5555\n"
	"01032321111\n"
	"hello send me a message(name.kwak@cau.ac.kr)\n"
	"call me (032 111 1424)")

subst = "\\1-****-\\2"

# You can manually specify the number of replacements by changing the 4th argument
result = re.sub(regex, subst, test_str, 0, re.MULTILINE)

if result:
    print (result)

# Note: for Python 2.7 compatibility, use ur"" to prefix the regex and u"" to prefix the test string and substitution.
```
